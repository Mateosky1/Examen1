```python
import math
alpha = 1 / math.log(10)

def briggs_method(x, n_roots=5, sig_figs=6):
    current_val = x
    for _ in range(n_roots):
        current_val = math.sqrt(current_val)
        if current_val != 0:
            power = sig_figs - 1 - int(math.floor(math.log10(abs(current_val))))
            current_val = round(current_val, power)
            
    r = current_val - 1
    approx_log = (2**n_roots) * alpha * r
    return round(approx_log, 6)

def calc_error(approx, exact):
    return abs((exact - approx) / exact)

exact_logs = {i: math.log10(i) for i in range(2, 10)}

log_2_aprox = briggs_method(2)
log_3_aprox = briggs_method(3)
log_7_aprox = briggs_method(7)

log_4_aprox = 2 * log_2_aprox
log_5_aprox = 1 - log_2_aprox
log_6_op1_aprox = log_2_aprox + log_3_aprox

r_6 = 1.0077696 - 1
log_base_6_op2 = alpha * r_6
log_6_op2_aprox = (log_base_6_op2 + 7) / 9

log_8_aprox = 3 * log_2_aprox
log_9_aprox = 2 * log_3_aprox

print("Logaritmo 2:", log_2_aprox, "| Error:", round(calc_error(log_2_aprox, exact_logs[2]), 6))
print("Logaritmo 3:", log_3_aprox, "| Error:", round(calc_error(log_3_aprox, exact_logs[3]), 6))
print("Logaritmo 4:", round(log_4_aprox, 6), "| Error:", round(calc_error(log_4_aprox, exact_logs[4]), 6))
print("Logaritmo 5:", round(log_5_aprox, 6), "| Error:", round(calc_error(log_5_aprox, exact_logs[5]), 6))
print("Logaritmo 6 (Op 1):", round(log_6_op1_aprox, 6), "| Error:", round(calc_error(log_6_op1_aprox, exact_logs[6]), 6))
print("Logaritmo 6 (Op 2):", round(log_6_op2_aprox, 6), "| Error:", round(calc_error(log_6_op2_aprox, exact_logs[6]), 6))
print("Logaritmo 7:", log_7_aprox, "| Error:", round(calc_error(log_7_aprox, exact_logs[7]), 6))
print("Logaritmo 8:", round(log_8_aprox, 6), "| Error:", round(calc_error(log_8_aprox, exact_logs[8]), 6))
print("Logaritmo 9:", round(log_9_aprox, 6), "| Error:", round(calc_error(log_9_aprox, exact_logs[9]), 6))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

alpha = 1 / math.log(10)

r_vals = np.linspace(0, 0.1, 500)

log_real = np.log10(1 + r_vals)
log_aprox = alpha * r_vals

error_absoluto = np.abs(log_real - log_aprox)
max_error = np.max(error_absoluto)

plt.figure(figsize=(10, 5))
plt.plot(r_vals, log_real, label="log10(1+r) Exacto", color="blue", linewidth=2)
plt.plot(r_vals, log_aprox, label="Aproximacion alpha * r", color="orange", linestyle="--", linewidth=2)
plt.title("Aproximacion Lineal de Briggs vs Valor Exacto")
plt.xlabel("Valor de r")
plt.ylabel("Resultado")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(r_vals, error_absoluto, label="Error Absoluto", color="red")
plt.title("Error Absoluto de la Aproximacion")
plt.xlabel("Valor de r")
plt.ylabel("Error")
plt.legend()
plt.grid(True)
plt.show()

print(f"El maximo error absoluto en el rango evaluado es: {max_error:.6f}")
```
