```python
import math

i = 0

def secant_method_optimized(f, x0, x1, tol=1e-4, max_iter=100):
    x_prev = x0
    x_curr = x1
    iter_count = 0
    
    f_prev = f(x_prev)
    f_curr = f(x_curr)
    
    while abs(f_curr) > tol and iter_count < max_iter:
        if f_curr - f_prev == 0:
            print("Division por cero detectada.")
            break
            
        x_next = x_curr - f_curr * (x_curr - x_prev) / (f_curr - f_prev)
        
        x_prev = x_curr
        x_curr = x_next
        
        f_prev = f_curr
        f_curr = f(x_curr) 
        
        iter_count += 1
        
    return x_curr, iter_count


def func_1(x):
    global i
    i += 1
    return x**3 + 2*x**2 + 3*x + 1

i = 0
raiz_1, iteraciones_1 = secant_method_optimized(func_1, x0=2, x1=3)
print(f"Prueba 1: Raiz = {raiz_1}, Llamadas a la funcion (i) = {i}")


# Prueba 2
def func_2(x):
    global i
    i += 1
    return 35 * math.sin(x / math.pi) + 10

i = 0 
raiz_2, iteraciones_2 = secant_method_optimized(func_2, x0=2, x1=3)
print(f"Prueba 2: Raiz = {raiz_2}, Llamadas a la funcion (i) = {i}")
```
