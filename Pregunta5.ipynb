Usando los métodos estudiados, determine todas las raíces de la ecuación:
𝑥5+5.25⋅𝑥4+4.125⋅𝑥3−9.125⋅𝑥2−14.625⋅𝑥−5.625=0.

```python
import numpy as np
from scipy import optimize

def funcion_polinomial(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625

def derivada_polinomial(x):
    return 5*x**4 + 21*x**3 + 12.375*x**2 - 18.25*x - 14.625

puntos_iniciales = [-4.0, -1.5, 2.0]

raices_encontradas = []

print("Calculando raices:")

for p0 in puntos_iniciales:
    try:
        raiz = optimize.newton(
            func=funcion_polinomial, 
            x0=p0, 
            fprime=derivada_polinomial, 
            tol=1e-6, 
            disp=False
        )
        raiz_redondeada = round(raiz, 4)
        if raiz_redondeada not in raices_encontradas:
            raices_encontradas.append(raiz_redondeada)
            
    except RuntimeError:
        print(f"El metodo no convergio para el punto inicial {p0}")

raices_encontradas.sort()

print("Raices unicas encontradas en orden ascendente:")
for i, r in enumerate(raices_encontradas):
    print(f"x_{i+1} = {r}")
```
